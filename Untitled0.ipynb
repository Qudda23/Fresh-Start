```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from google.colab import drive
drive.mount('/content/drive')

# Load dataset
data = pd.read_csv('/content/drive/MyDrive/telecom_netone.csv')

# Display dataset
print(data.head())

# Features and target
X = data[['Subscribers_Millions', 'Market_Share_Percent']]
y = data['Data_Usage_TB']

# Train model
model = LinearRegression()
model.fit(X, y)

# Predictions
data['Predicted_Data_Usage'] = model.predict(X)

# Plot actual vs predicted
plt.figure()
plt.plot(data['Year'], data['Data_Usage_TB'], label='Actual Data Usage')
plt.plot(data['Year'], data['Predicted_Data_Usage'], label='Predicted Data Usage')
plt.xlabel('Year')
plt.ylabel('Data Usage (TB)')
plt.legend()
plt.title('NetOne Data Usage Prediction Model')
plt.show()

# Correlation
corr = data.corr(numeric_only=True)
print(corr)

# Heatmap
plt.figure()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix of NetOne Telecom Data")
plt.show()

# Predict future scenario
future_data = pd.DataFrame({
    'Subscribers_Millions': [5.2],
    'Market_Share_Percent': [24]
})

prediction = model.predict(future_data)
print("Predicted Future Data Usage:", prediction)

# Create future years dataset
future_years = pd.DataFrame({
    'Year': [2025, 2026, 2027],
    'Subscribers_Millions': [5.2, 5.6, 6.0],   # projected growth
    'Market_Share_Percent': [23.5, 24.0, 24.5]
})

# Predict future data usage
future_years['Predicted_Data_Usage'] = model.predict(
    future_years[['Subscribers_Millions', 'Market_Share_Percent']]
)

print(future_years)

# Combine datasets
combined = pd.concat([
    data[['Year', 'Data_Usage_TB']],
    future_years[['Year', 'Predicted_Data_Usage']].rename(columns={'Predicted_Data_Usage':'Data_Usage_TB'})
])

plt.figure()

# Historical
plt.plot(data['Year'], data['Data_Usage_TB'], label='Historical Data')

# Forecast
plt.plot(future_years['Year'], future_years['Predicted_Data_Usage'], linestyle='dashed', label='Forecast (2025-2027)')

plt.xlabel('Year')
plt.ylabel('Data Usage (TB)')
plt.title('NetOne Data Usage Forecast to 2027')
plt.legend()
plt.show()
```
